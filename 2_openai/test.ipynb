```python
import asyncio
import time

# --- The Sequential Line-by-Line Workflow ---
async def generate_user_forecast(user_name):
    print(f"[{time.strftime('%X')}] 👤 {user_name}: Step 1 - Fetching database actuals...")
    
    # The workflow pauses completely right here. It cannot move to Step 2.
    # But it tells the server: "Go help someone else while I wait."
    await asyncio.sleep(4)
    
    print(f"[{time.strftime('%X')}] 👤 {user_name}: Step 2 - Calculating new forecast...")
    
    # Pauses completely again.
    await asyncio.sleep(2) 
    
    print(f"[{time.strftime('%X')}] ✅ {user_name}: Forecast Complete!")


# --- The Server Simulation ---
async def main():
    print(f"[{time.strftime('%X')}] --- Server Online ---")
    
    # User A clicks the button. The server starts their workflow in the background.
    task_a = asyncio.create_task(generate_user_forecast("User A"))

    await asyncio.sleep(5) 
    # await task_b
    # 1 second later, User B logs in and clicks the button.
    task_b = asyncio.create_task(generate_user_forecast("User B"))

    await asyncio.sleep(3)
     
    # Keep the server running until both are done
    # await task_a
    await task_b
    await task_a

await main()
```
